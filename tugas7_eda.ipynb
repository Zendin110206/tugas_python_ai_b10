### 1. Image 1: The Three Spikes (Why does it look broken?)
Look at my first image with the blue, red, and green charts. It looks like one giant tower on the far left and then a flat line to the right.
* **What is happening:** This is called a **Right-Skewed Distribution**.
* **The "Why":** Most countries are relatively small and have lower case numbers. However, a tiny handful of countries (like the USA, India, or Brazil) have *millions* of cases.
* **The Problem:** To fit the USA's 5,000,000 cases on the screen, matplotlib had to stretch the X-axis extremely far to the right. Because of that stretch, all the "normal" countries get smashed together into that single tall spike on the left. The chart is mathematically correct, but it is completely useless for analysis because we can't see the details of the normal countries.

> Additional notes:
> * **`1e6`** is scientific notation for $1 \times 10^6$. That means **1 Million**. ( 1.2 which is 12 million).

**What is the chart telling us?**

**4. Analisis Tes dan Positivity Rate**
> Buat kolom baru "Positivity Rate" = TotalCases / TotalTests

> Analisis negara-negara dengan positivity rate tertinggi

> i want to know why doing this

### The "Why": What is a Positivity Rate?
Imagine two countries: **Country A** and **Country B**.
* They both report **1,000 Total Cases** today.
* If you only look at Total Cases, they look exactly the same, right?

But wait. Let's look at their testing:
* **Country A** only tested **2,000** people to find those 1,000 cases. That means **50%** of everyone they tested was sick. This is terrifying. The virus is everywhere, and they are only testing the people who are already at the hospital dying.
* **Country B** tested **1,000,000** people to find those 1,000 cases. That means only **0.1%** of the tests were positive. This country is extremely safe; they are testing randomly, and hardly anyone is sick.

**Positivity Rate = (Total Cases ÷ Total Tests) × 100.** It is the ultimate lie-detector in epidemiology. It tells us which countries are hiding massive outbreaks because they refuse to test enough people. We are hunting for the countries with the highest percentages.

```python
df['Positivity Rate (%)'] = (df['TotalCases'] / df['TotalTests']) * 100
top_positivity = df.dropna(subset=['Positivity Rate (%)']).sort_values(by='Positivity Rate (%)', ascending=False).head(10) # dropna biar gak crash, lalu di sorting dari highest

print("=== Top 10 Countries by Positivity Rate ===")
display(top_positivity[['Country/Region', 'TotalCases', 'TotalTests', 'Positivity Rate (%)']])
```

This data is not correct i believe, i mean how could the total cases is 10.000+ but the total test is just 401.0??

More explanation:

### The Problem: Impossible Math
Look closely at the row for **Sudan**:
* **Total Cases:** 11,780
* **Total Tests:** 401

The formula for Positivity Rate is `(Total Cases / Total Tests) * 100`.
If you do `(11,780 / 401) * 100`, you get exactly **2,937.65%**.

**Why is this impossible?**
You cannot physically have 11,780 positive cases if you only tested 401 people! The absolute maximum number of positive cases from 401 tests is 401 (which would be a 100% positivity rate).

### Why does real-world data do this?
When you work with global datasets (like Worldometer), you are relying on governments to report accurate numbers. This anomaly happens because of:
1. **Data Entry Errors:** The person typing the data for Sudan might have missed a few zeros on the "Total Tests" column.
2. **Clinical Diagnoses vs. Lab Tests:** A doctor might look at a patient, say "You definitely have COVID," and count them as a "Case" without ever using an official laboratory "Test."


> My solution

```python
df['Positivity Rate (%)'] = (df['TotalCases'] / df['TotalTests']) * 100

# DATA CLEANING: Removing impossible values
valid_tests_df = df[df['Positivity Rate (%)'] <= 100]

top_positivity = valid_tests_df.sort_values(by='Positivity Rate (%)', ascending=False).head(10)

print("=== Top 10 Countries by VALID Positivity Rate ===")
display(top_positivity[['Country/Region', 'TotalCases', 'TotalTests', 'Positivity Rate (%)']])
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(data=top_positivity, x='Positivity Rate (%)', y='Country/Region',
            hue='Country/Region', palette='Blues_r', legend=False) # pakai Blues_r biar bagus

plt.title('Top 10 Countries with the Highest Positivity Rate', fontsize=14, fontweight='bold')
plt.xlabel('Positivity Rate (%)', fontsize=12)
plt.ylabel('Country / Region', fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**5. Death Rate & Recovery Rate**
> Buat kolom baru "Death Rate" = TotalDeaths / TotalCases

> Buat kolom "Recovery Rate" = TotalRecovered / TotalCases

> Bandingkan death rate dan recovery rate secara global

This step is incredibly important because it shifts our focus from how fast the virus is spreading to how successfully the world is surviving it.

```python
df['Death Rate (%)'] = (df['TotalDeaths'] / df['TotalCases']) * 100
df['Recovery Rate (%)'] = (df['TotalRecovered'] / df['TotalCases']) * 100

global_cases = df['TotalCases'].sum()
global_deaths = df['TotalDeaths'].sum()
global_recovered = df['TotalRecovered'].sum()

global_death_rate = (global_deaths / global_cases) * 100
global_recovery_rate = (global_recovered / global_cases) * 100

print("=== Global Outcome Statistics ===")
print(f"Global Recovery Rate: {global_recovery_rate:.2f}%")
print(f"Global Death Rate: {global_death_rate:.2f}%")
```

> If we sum both of them it is not 100%, i believe because there are still active case?

```python
plt.figure(figsize=(8, 5))

rates = [global_recovery_rate, global_death_rate]
labels = ['Global Recovery Rate', 'Global Death Rate']

sns.barplot(x=labels, y=rates, hue=labels, palette=['#2ECC71', '#E74C3C'], legend=False)

plt.title('Global Comparison: Recovery vs Death Rate', fontsize=14, fontweight='bold')
plt.ylabel('Percentage (%)', fontsize=12)
plt.ylim(0, 100)

# Adding the exact percentage text on top of the bars
for i, v in enumerate(rates):
    plt.text(i, v + 2, f"{v:.2f}%", ha='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='Recovery Rate (%)', y='Death Rate (%)',
                alpha=0.7, color='purple', edgecolor='black')

plt.title('Global Distribution: Recovery Rate vs Death Rate by Country', fontsize=14, fontweight='bold')
plt.xlabel('Recovery Rate (%)', fontsize=12)
plt.ylabel('Death Rate (%)', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Based on the comprehensive Exploratory Data Analysis of the Worldometer dataset, the most significant finding is not just the behavior of the virus, but the severe flaws within the data itself.
